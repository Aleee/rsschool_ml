### Simple version (1 point)

```python
# 1 point
def scale(X):
    return (X - np.mean(X, axis=0))/np.std(X)
```

```python
assert np.allclose(np.array(scale(wine_data)), StandardScaler().fit_transform(wine_data))
```

### Pipeline Version (2 points)

```python
# 2 points
from sklearn.base import BaseEstimator, TransformerMixin
class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self, copy=True, with_mean=True, with_std=True):
        self.copy = copy
        self.with_mean = with_mean
        self.with_std = with_std

    def fit(self, X, y=None):
        self.X_mean = int(self.with_mean) * np.mean(X, axis=0)
        self.X_std = int(self.with_std) * np.std(X) + int(not self.with_std) * 1
        return self

    def transform(self, X, y=None, copy=None):
        return (X - self.X_mean) / self.X_std
```

```python
assert np.allclose(CustomScaler().fit_transform(wine_data), StandardScaler().fit_transform(wine_data))
```

## Exercise  3 - Preprocessing Pipeline (2 points)

```python
wine_train, wine_val, wine_labels_train, wine_labels_val = train_test_split(wine_data, wine_labels, 
                                                                            test_size=0.3, random_state=1)
```

Train model (for example, `LogisticRegression(solver='liblinear', penalty='l1')` on raw `wine_train` data; then train same model after data scaling; then add feature selection (and train model again on scaled data). For each experiment all required preprocessing steps (if any) should be wrapped into sklearn pipeline.

Measure `accuracy` of all 3 approaches on `wine_val` dataset. Describe and explain results.

```python
def make_model(scale=False, feat_sel=False):
    pipeline = []
    if scale:
        pipeline.append(('sca', StandardScaler()))
    if feat_sel:
        pipeline.append(('sel', SelectFromModel(estimator=LogisticRegression())))
    pipeline.append((('reg', LogisticRegression(solver='liblinear', penalty='l1', C=1.5))))
        
    return Pipeline(pipeline)

def measure_accuracy(model_func):
    return accuracy_score(wine_labels_val, model_func.fit(wine_train, wine_labels_train).predict(wine_val))
```

```python
results = pd.DataFrame(columns=['Accuracy'])
results.loc["Logistic regression"] = measure_accuracy(make_model())
results.loc["Logistic regression with scaler"] = measure_accuracy(make_model(scale=True))
results.loc["Logistic regression with scaler and feature selection"] = measure_accuracy(make_model(scale=True, feat_sel=True))
results
```

Actually, we had to try several feature selection methods and regularization strengths in order to achieve such a result: scaling and feature selection consecutively increase model's accuracy. Many setups would offer no accuracy gain following these steps or even a decrease (that is especially true for feature selection whuch may be simply unfortunate). Therefore preprocessing is rather an instrument for model enhancement, not a magic stick.

## Exercise 4 - manual PCA (5 points)
The task is to solve PCA as an optimization problem, without explicitly doing eigen value decomposition.
In the most general setting PCA is minimization of reconstruction error of a projection of given rank $q$

$$\min_{\mu, \lambda_1,\ldots, \lambda_n, \mathbf{V}_q} \sum_{i=1}^n ||x_i - \mu - \mathbf{V}_q \lambda_i||^2$$

With a number of steps that can be found here https://stats.stackexchange.com/a/10260 this task transforms to
 $$\max_{u_i} \sum_{i=1}^q u_i^T \mathbf{S} u_i$$
 where $\mathbf{S}$ is the sample covariance matrix (after standartization) and $u_1, \ldots, u_q$ are the $q$ are orthonormal columns in $\mathbf{V}_q$.
 Let us solve this optimization problem with `scipy.optimize` library.
 
 Additional 2 point are given for visualization of the results.

### PCA (3 points)

```python
wine_data, wine_labels = wine_sklearn['data'], wine_sklearn['target']
```

Find a covariance matrix of standartized data and assing it to S.

```python
S = np.cov(StandardScaler().fit_transform(wine_data), rowvar=False)
np.linalg.norm(S)
```

If your code is correct, the following assert should be Ok.

```python
assert np.allclose(np.linalg.norm(S), 5.787241159764733)
```

```python
from scipy.optimize import minimize

def objective(x):
    # your code: write objective of the problem (don't forget that scipy does min while we need max)
    return -1 * x.T @ S @ x

def norm_constraint(x):
    # your code: constaraint norm of x to be 1, function should return 0 if constraint holds
    if np.linalg.norm(x) != 1:
        return 1 - x.T @ x
    else:
        return 0

con1 = {'type': 'eq', 'fun': norm_constraint}

x0 = np.zeros(wine_data.shape[1])

sol = minimize(objective, x0, constraints = [con1])
x0 = sol.x
```

Hurray! We have first vector! Let's do another one.

```python
def orthogonality_constraint(x):
    # your code: x should be orthogonal to x0, function should return 0 if constraint holds
    if x.T @ x0 != 0:
        return -1 * x.T @ x0
    else:
        return 0

con2 = {'type': 'eq', 'fun': orthogonality_constraint}

x1 = np.zeros(wine_data.shape[1])

sol = minimize(objective, x1, constraints = [con1, con2])
x1 = sol.x
```

If your solution is correct, the following asserts should be Ok.

```python
assert np.allclose(x0@S@x0, 4.732436977583595)
assert np.allclose(x1@S@x1, 2.5110809296451233)
```

Visualize the points after applying custom dimension reduction with 2 components.
